# Train test split

```python
x = cust.drop(columns = 'Exited')
y = cust['Exited']
```

```python
xtrain,xtest,ytrain,ytest = train_test_split(x,y,train_size=.8,stratify=y,random_state=1)
```

```python
print(xtrain.shape,ytrain.shape,xtest.shape,ytest.shape)
```

### Outliers Treatment

```python
# Cap the outliers using IQR Method
for i in num_cols:
    q3, q1 = xtrain[i].quantile([0.75,0.25])
    iqr = q3 - q1
    ul = q3 + 1.5 * iqr
    ll = q1 - 1.5 * iqr
    xtrain[i] = xtrain[i].apply(lambda x : ul if x > ul else ll if x < ll else x)
```

```python
t = 1
plt.figure(figsize=(11,7))
plt.suptitle('Univariate (Numerical)')
for i in num_cols:
    plt.subplot(2,2,t)
    sns.boxplot(xtrain, x = i)
    t+=1
plt.tight_layout()
plt.show()
```

```python
xtrain.duplicated().sum()
```

## Feature Engineering

### Encoding

```python
xtrain.head(3)
```

```python
map_dic = round(xtrain['Geography'].value_counts(normalize=True),4).to_dict()
map_dic
```

```python
xtrain['Geography'] = xtrain['Geography'].map(map_dic)
xtest['Geography'] = xtest['Geography'].map(map_dic)
```

```python
xtest.head()
```

```python
xtrain.head()
```

```python
xtrain = pd.get_dummies(xtrain, columns=['Gender'],dtype=int,drop_first=True)
xtest = pd.get_dummies(xtest, columns=['Gender'],dtype=int,drop_first=True)
```
